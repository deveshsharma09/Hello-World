```python
class Shape():
  def area(self):
    pass                                  #This is an example of inheritance and polymorphism in oops.
                                          #Here shape is main class
  def perimeter(self):
    pass


class Square(Shape):                       #Square, rectangle and circle are sub classes which inehrit the properties of main class Shape
  def __init__(self, side):
    self.side = side

  def area(self):                           #This is an example of polymorphism where we are changing some properties of a method called area defined in main class.
    return self.side**2

  def perimeter(self):
    return self.side*4


class Rectangle(Shape):
  def __init__(self, height, width):
    self.height = height
    self.width = width

  def area(self):
    return self.height*self.width

  def perimeter(self):
    return 2*(self.height+self.width)



class Circle(Shape):
  def __init__(self, radius):
    self.radius = radius
    self.pie = 3.14

  def area(self):
    return 2*self.radius*self.pie

  def perimeter(self):
    return self.pie*self.radius**2


```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data (replace with your actual data)
order_data = {
    'Product Category': ['Office Supplies', 'Furniture', 'Technology'],
    'High': [50, 30, 40],  # Number of orders with high priority
    'Medium': [100, 80, 70],  # Number of orders with medium priority
    'Low': [80, 60, 50]  # Number of orders with low priority
}

# Convert data to DataFrame
df = pd.DataFrame(order_data)

# Plotting
plt.figure(figsize=(10, 6))

# Stacked bar chart
bottom = None
for priority in ['High', 'Medium', 'Low']:
    plt.bar(df['Product Category'], df[priority], label=priority, bottom=bottom)
    if bottom is None:
        bottom = df[priority].values
    else:
        bottom += df[priority].values

plt.title('Distribution of Order Priorities Across Product Categories')
plt.xlabel('Product Category')
plt.ylabel('Number of Orders')
plt.legend(title='Priority')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd

# Sample data (replace with your actual data)
region_returns = {
    'Region': ['Region A', 'Region B', 'Region C'],
    'Returns': [100, 150, 80],  # Number of returned orders per region
    'Total Sales': [10000, 12000, 9000]  # Total sales per region
}

# Convert data to DataFrame
df_returns = pd.DataFrame(region_returns)

# Calculate proportion of returns
df_returns['Return Rate'] = df_returns['Returns'] / df_returns['Total Sales']

# Load geographical data (shapefile for regions)
# Replace 'path_to_shapefile' with the actual path to your shapefile
gdf = gpd.read_file('path_to_shapefile')

# Merge geographical data with returns data
merged = gdf.merge(df_returns, how='left', left_on='Region', right_on='Region')

# Plotting - Choropleth Map
fig, ax = plt.subplots(1, 1, figsize=(10, 8))
merged.plot(column='Return Rate', cmap='OrRd', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True)
ax.set_title('Geographical Distribution of Returns')
plt.show()

# Plotting - Bar Chart
plt.figure(figsize=(10, 6))
plt.bar(df_returns['Region'], df_returns['Total Sales'], label='Total Sales', color='skyblue')
plt.bar(df_returns['Region'], df_returns['Returns'], label='Returned Orders', color='salmon', alpha=0.7)
plt.xlabel('Region')
plt.ylabel('Amount')
plt.title('Profitability by Region with Returns Impact')
plt.legend()
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
# 1. Handle Missing Values
#df.dropna(inplace=True)  # Drop rows with missing values
df.fillna(df.mean(), inplace=True)
df.fillna(df.mode().iloc[0], inplace=True)

# 2. Convert Data Types
#df['Date'] = pd.to_datetime(df['Date'])  # Convert Date column to datetime

# 3. Handle Duplicates
df.drop_duplicates(inplace=True)  # Remove duplicate rows

# 4. Feature Engineering (if applicable)
# For example, extract year and month from the Date column
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month

from sklearn.preprocessing import LabelEncoder, MinMaxScaler
label_encoder = LabelEncoder()
for column in df.select_dtypes(include=['object']).columns:
    df[column] = label_encoder.fit_transform(df[column])

# 5. Normalize or Scale Data (if applicable)
# For example, use Min-Max scaling on numerical columns
#from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df[['Open', 'High', 'Low', 'Close']] = scaler.fit_transform(df[['Open', 'High', 'Low', 'Close']])

# 6. Handle Outliers (if applicable)
# For example, winsorize outliers in the 'Close' column
from scipy.stats import mstats
df['Close'] = mstats.winsorize(df['Close'], limits=[0.05, 0.05])

# Display the modified DataFrame
print(df.head())
```

```python

import matplotlib.pyplot as plt

# Sort the DataFrame by the 'Date' column
df_sorted = df.sort_values(by='Date')

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(df_sorted['Date'], df_sorted['Close'], label='Closing Price', color='blue')
plt.plot(df_sorted['Date'], df_sorted['Open'], label='Opening Price', color='red')
plt.title('Close vs Open Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
